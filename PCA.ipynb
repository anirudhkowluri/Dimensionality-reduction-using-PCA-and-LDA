```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
y = pd.DataFrame(cancer['target'])
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaled_features = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)#  we will get  10  best components#
pca.fit(scaled_features)
x_pca = pca.transform(scaled_features)
x_pca.shape
```

```python
pca.explained_variance_ratio_ #covariance matrix of all the features or covariance percentageof top 10 features if we add them we get 95%#
```

```python
sum(pca.explained_variance_ratio_[:10])*100 #covariance matrix of all the features percentage explained by first 10 components#
```

```python
# plot the covariance ratio of the first 10 principal components#
plt.figure(figsize=(15,10))
sns.heatmap(pd.concat([pd.DataFrame(x_pca),y], axis=1).corr(), annot=True,fmt='0.3f')
plt.show()
```

```python
# now train the model with the scaled data
X_train, X_test, y_train, y_test = train_test_split(x_pca,y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
svc1 = SVC()
svc1.fit(X_train, y_train)
y_pred1 = svc1.predict(X_test)
accuracy_score(y_test,y_pred1)
```

```python
# after using pca the accuracy is increased from 95 to 97 and the no of features are also reduced 
```
